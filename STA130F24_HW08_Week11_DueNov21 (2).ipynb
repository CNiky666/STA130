(a)A Classification Decision Tree addresses problems where the goal is to predict the class or category of an observation based on input features. It works by segmenting the feature space into regions based on thresholds in the data and assigning a class label to each region. The tree structure allows for straightforward decision-making, making it interpretable and easy to use.

Type of Problem
A Classification Decision Tree is typically used for supervised learning problems where:

Input data: A set of features (independent variables) are given.
Output: The desired outcome is a categorical label (e.g., "Yes" or "No", "Class A" or "Class B").
Key Features of Problems Suitable for Decision Trees
Non-linear relationships between variables.
Data with missing or irrelevant features.
Need for an interpretable model.
Problems where decision rules are useful for domain experts.
Real-World Applications
Medical Diagnosis

Predicting whether a patient has a specific disease based on symptoms, lab results, or imaging data.
Example: Classifying patients as diabetic or non-diabetic based on glucose levels, age, and other health metrics.
Customer Segmentation

Identifying different types of customers for targeted marketing campaigns.
Example: Predicting whether a customer will respond to a promotion based on demographics and past purchase behavior.
Fraud Detection

Classifying transactions as fraudulent or legitimate based on transaction data such as amount, location, and time.
Example: Detecting fraudulent credit card transactions.

(b)1. Nature of the Problem
Classification Decision Tree:

Solves classification problems, predicting categorical outcomes (e.g., "Yes" or "No," "A," "B," or "C").
Example: Predicting whether a customer will churn or not.
Multiple Linear Regression:

Solves regression problems, predicting continuous numerical values (e.g., house prices, income).
Example: Predicting house prices based on size, number of rooms, and location.
2. How Predictions Are Made
Classification Decision Tree:
Structure:

The model builds a tree structure with decision nodes and leaf nodes. Decision nodes evaluate conditions on input features (e.g., "Is age greater than 30?").
Each leaf node corresponds to a predicted class (e.g., "Spam" or "Not Spam").
Process:

Start at the root of the tree.
At each decision node, evaluate the condition based on the feature values of the input.
Follow the branch corresponding to the outcome of the condition.
Continue until reaching a leaf node, which gives the predicted class.
Output:

The predicted class label is determined by the majority class at the leaf node.
Alternatively, probabilities of each class can be provided based on the proportion of observations in the training data that fall into each class at the leaf.
Multiple Linear Regression:
Structure:

The model assumes a linear relationship between the independent variables (features) and the dependent variable (target). The prediction is based on a linear equation:
y = beta_0 + beta_1 * x1 + beta_2 * x2 + ... + beta_n * xn + epsilon
Here:
y is the predicted value (continuous).
x1, x2, ..., xn are the input features.
beta_0 is the intercept (constant term).
beta_1, beta_2, ..., beta_n are the coefficients (weights) of the features.
epsilon is the error term (unexplained variance).
Process:

Multiply each feature by its corresponding coefficient.
Sum these products along with the intercept.
The result is the predicted value, y.
Output:

A single continuous numerical value, representing the prediction.

1. **Accuracy** measures the proportion of true results (both true positives and true negatives) in the population.

$$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$
    
2. **Sensitivity** measures the proportion of actual positives that are correctly identified.

$$\text{Sensitivity} = \frac{TP}{TP + FN}$$

3. **Specificity** measures the proportion of actual negatives that are correctly identified.

$$\text{Specificity} = \frac{TN}{TN + FP}$$

4. **Precision** measures the proportion of positive identifications that were actually correct.

$$\text{Precision} = \frac{TP}{TP + FP}$$

Scenario: Spam email filtering.

> The way a **classification decision tree** is fit is that at each step in the construction process of adding a new **decision node splitting rule** to the current **tree structure**, all possible **decision rules** for all possible **predictor variables** are considered, and the combination that improves the **prediction** the most (as measured by the criterion of either "Gini impurity" or "Shannon entropy") and in accordance with the rules of the decision tree (such as the `max_depth` argument) is added to the **classification decision tree**. But the so-called **feature importance** heuristics allows us to judge how relatively important the overall contributions from different features are in the final decision tree predictions.
