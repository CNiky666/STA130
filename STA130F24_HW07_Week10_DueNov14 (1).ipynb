> _The way an **interaction** actually appears and works in the context of the linear form of a **multiple linear regression** model may not necessarily be immediately intuitive, as it is actually in fact somewhat subtle and tricky. Of course, an **interaction** is when the relationship of one **predictor variable** with the **outcome variable** depends on the value of another different **predictor variable**, so the impact of one **predictor variable** changes based on the presence or magnitude of another **predictor variable**. But are you sure you know what this means in the context of the **linear form** of a **multiple linear regression** model?_
>
> - _Imagine you're selling smoothies, the taste of your smoothie depends on the ingredients included in the smoothie, and there are two ingredients (bananas and strawberries) in the smoothie._
>     - _Adding more bananas into the smoothie will of course increase the "banana" flavor of the smoothie, and vice-versa for strawberries..._
>     - _But is this "banana" influence on the flavor always constant, or can it change depending on the absolute amount of strawberries in the smoothie?_ 
>     - _If the "banana" flavor influence is constant and does not depend on the  absolute amount of strawberries in the smoothie, then there is no **interaction** and the **linear form** of the model is $\beta_b b_i + \beta_s s_i$ and the model is said to be only **additive**._
>     - _But if the "banana" flavor influence does depend on the absolute amount of strawberries in the smoothie, then there IS an **interaction** and the **linear form** of the model is $\beta_b b_i + \beta_s s_i + \beta_{bs} (b_i\times s_i)$ and the model is said to be **synergistic**._
>         
> _These **linear forms** show that either bananas and strawberries do not have any **synergistic interaction** and contribute to the flavor independently; or, they do have a **synergistic interaction** and there is an interesting interplay between bananas and strawberries in the way they influence the taste of the smoothie._ 
> 
> - _So, if there is no **interaction**, then the effect of adding more bananas on the taste of the smoothie will always be the same, no matter how many strawberries you put in. So the effect of bananas on the smoothie is the same whether you add a lot of strawberries or just a few: $\beta_b b_i + \beta_s s_i$_
> - _Or, on the other hand, if there is an **interaction**, then the effect of adding bananas (on the smootie flavor) will be different depending on how many strawberries there currently are in the smoothie: $\beta_b b_i + \beta_s s_i + \beta_{bs} (b_i\times s_i)$_
> 
> _In this case, the right answer is probably that the **linear form** with the **interaction** is correct. This is because the flavor probably depends on the relative amount of bananas and strawberries in the smoothie; so, the effect of adding a fixed amount of bananas to the smoothie probalby depends on the absolute amount of strawberries that are in the smoothie._
> 
> _Again, because understanding **interactions** in the context of **linear forms** is somewhat subtle and tricky and indeed not necessarily obviously intuitive, let's think about this a bit more. And we can simplify the concept a little bit by considering how this **interaction** would actually technically work in a **linear form** if we just had **binary indicator variables**._
>         
> - _To consider the smootie example in terms of binary variables, suppose that if both fruits are added to the smootie, they will be added in the same amount. So the smoothie will be made with either just bananas, just strawberries, or both (or neither and you won't make a smoothie)._ 
>     - _The question regarding an **interaction** then is, is the influence of the ingredients on the taste of the smoothie **additive** or **synergistic**? That is, does the way bananas affects the flavor of the smoothie change depending on the inclusion or exclusion of strawberries in the smoothie?_
>     - _**Additive** $\beta_b 1_{[b_i=1]}(b_i) + \beta_s 1_{[s_i=1]}(s_i)$ means there are three different flavors but they are explained by just two **parameters**: banana $\beta_b$, strawberry $\beta_s$, and banana-strawberry $\beta_b+\beta_s$_
>     - _**Synergistic** $\beta_b 1_{[b_i=1]}(b_i) + \beta_s 1_{[s_i=1]}(s_i) + \beta_{bs}(1_{[b_i=1]}(b_i) \times 1_{[s_i=1]}(s_i))$ means there are of course again three different flavors, but this time they are explained by three **parameters**: banana $\beta_b$, strawberry $\beta_s$, and banana-strawberry $\beta_b+\beta_s + \beta_{bs}$, which indicates that the flavor is "more than just sum of its parts", meaning there is a **synergistic interaction** and there is an interesting interplay between bananas and strawberries in the way they influence the taste of the smoothie_
>     
> _As the **additive** and **synergistic** versions of the **linear form** of the two **binary indicator variables** context shows, we don't need an interaction to make different predictions for different combinations of things. Instead, what these show is that the prediction will either be **additive** and "just the sum of it's parts" or **synergistic** (**interactive**) and "more than just sum of its parts"._

In this scenario, we have the following variables:

Outcome Variable: Sales (the amount of sports equipment sold).
Predictor Variables: TV advertising budget and online advertising budget. Both are continuous variables in the first part and are categorized as either "high" or "low" (binary variables) in the second part.
Step 1: Without Interaction (Two Continuous Variables)
Linear Form (without interaction):
outcome = β0 + β1 TV_budget + β2 online_budget
Explanation: This model assumes that the effect of each type of advertising is independent of the other. It estimates the direct impact of TV and online advertising budgets on sales separately.
Step 2: With Interaction (Two Continuous Variables)
Linear Form (with interaction):
outcome = β0 + β1 TV_budget + β2 online_budget + β3 (TV_budget * online_budget)
Explanation: This model includes an interaction term between the TV and online advertising budgets. It allows us to model the possibility that the effectiveness of spending in one channel depends on the spending in the other channel. For example, the impact of TV advertising might be greater when more is spent online and vice versa.
Step 3: Making Predictions
Without Interaction: The predicted outcome depends only on the separate effects of TV and online advertising budgets. For example, increasing the TV budget would lead to an increase in sales according to its own coefficient (β1), without considering how the online budget influences this relationship.
With Interaction: When making predictions using the model with the interaction term, the effect of changing one budget depends on the level of the other budget. The presence of the interaction term (β3) implies that the impact of each type of advertisement is not independent, and it could be amplified or diminished based on the other advertising medium's budget.
Step 4: Without Interaction (Two Binary Variables)
Linear Form (without interaction):
outcome = β0 + β1 1(TV_high) + β2 1(online_high)
Explanation: Here, the two advertising budgets are treated as binary variables ("high" or "low"). The coefficients β1 and β2 represent the effect on sales if the TV or online advertising budget is high, compared to a low budget, respectively.
Step 5: With Interaction (Two Binary Variables)
Linear Form (with interaction):
outcome = β0 + β1 1(TV_high) + β2 1(online_high) + β3 (1(TV_high) * 1(online_high))
Explanation: This model includes an interaction term between the two binary variables, indicating that the combined effect of high TV and high online advertising may be different from the sum of their individual effects. For example, having both high TV and high online might lead to an additional boost in sales that is not captured by the individual effects alone.
Step 6: Making Predictions with Binary Variables
Without Interaction: The prediction of sales is based on the additive effects of high or low TV and online advertising budgets. For example, if both TV and online budgets are high, the predicted outcome would be:
outcome = β0 + β1 + β2
With Interaction: When including the interaction, the prediction considers not just the individual effects of high advertising budgets but also the additional impact of having both high TV and online budgets:
outcome = β0 + β1 + β2 + β3
The interaction term (β3) represents the extra effect of having both high TV and high online budgets compared to just having them independently.
Summary
Model without Interaction: This is simpler and assumes that each predictor independently influences sales. It provides a linear, additive interpretation.
Model with Interaction: This model captures the idea that the effect of one predictor may depend on the level of the other predictor. It provides a more nuanced understanding, especially when predictors may influence each other’s impact. The interaction term allows the model to account for relationships between predictors that could lead to more accurate predictions of sales when they work in tandem.

Explain the apparent contradiction between the factual statements regarding the fit below that "the model only explains 17.6% of the variability in the data" while at the same time "many of the *coefficients* are larger than 10 while having *strong* or *very strong evidence against* the *null hypothesis* of 'no effect'"<br>

<details class="details-example"><summary style="color:blue"><u>Further Guidance</u></summary>
    
> _How do we simultaneously interpret **hypothesis testing** results regarding **coefficient estimates** based on **p-values** and **R-squared** "the proportion of variation in (outcome) $y$ explained by the model ($\hat y_i$)"? How can both be meaningfully understood at the same time? Do they address different aspects of a model?_
>    
> _As introduced in the previous homework, **R-squared** is_
>
> _$$R^2 = 1 - \frac{\sum_{i=1}^n(y_i-\hat y)^2}{\sum_{i=1}^n(y_i-\bar y)^2}$$_
>    
> _which describes the **explanatory power** of a model; whereas, **p-values** allow us to characterize **evidence against** a **null hypothesis**, and **coefficients** in a **multiple linear regression** context allow us to interpret the relationship between the **outcome** and a **predictor variable** "with all other **predictor variables** 'held constant'". Are these concepts thus contradictory or conflictual in some manner?_

|p-value|Evidence|
|-|-|
|$$p > 0.1$$|No evidence against the null hypothesis|
|$$0.1 \ge p > 0.05$$|Weak evidence against the null hypothesis|
|$$0.05 \ge p > 0.01$$|Moderate evidence against the null hypothesis|
|$$0.01 \ge p > 0.001$$|Strong evidence against the null hypothesis|
|$$0.001 \ge p$$|Very strong evidence against the null hypothesis|
    
> _In `formula='HP ~ Q("Sp. Def") * C(Generation)'` the `Q` stands for "quote" and is needed to access column names when they have a "space" in their name, while the `C` indicates a **categorical** use of what is actually an **integer** valued column. Despite technically being **continuous** numbers, **integer** often simply indicate categories which should not necessarily be treated as an incremental **continuous predictor variable**. Remember, a model such as $\beta_0 + \beta_1 x$ means for each unit increase in $x$ the outcome increases "on average" by $\beta_1$; so, if $x$ takes on the values `1` through `6` as the `Generation` **predictor variable** here does, then this means the average value for "Generation 1" must be $\beta_0 + \beta_1$ while for "Generation 2" it must be $\beta_0 + 2\times \beta_1$ (and so on up to "Generation 6" which must be $\beta_0 + 6\times \beta_1$). This might be a very strange restriction to place on something that is really actually a **categorical predictor variable**. You can see in the given model fit below how this six-level **categorical predictor variable** is actually appropriately treated in the specification of the **linear form** using "Generation 1" for the "baseline" and **binary indicators** to model the "contrast" ("offsets") for the other "Generations"; and, how these are in turn used in the context of the **interaction** considered by the model specification._

There is no inherent contradiction between the statements about R-squared and the p-values of the coefficients. They reflect different aspects of the model’s performance and statistical findings. Let’s address this in detail:

R-squared and Its Interpretation
R-squared (R²) represents the proportion of the variance in the outcome variable (y) that is explained by the model. It ranges from 0 to 1, where higher values indicate that the model explains more of the variability in the data.
In this case, R² = 0.176 indicates that only 17.6% of the variation in the data is explained by the model. This suggests that the model, overall, has limited explanatory power, and there is a substantial amount of unexplained variability in the outcome.
P-values and Coefficients
Coefficients represent the effect of each predictor variable on the outcome, assuming all other predictor variables are held constant. A larger coefficient indicates a stronger effect of that predictor on the outcome.
P-values are used to test the significance of these coefficients. A small p-value (typically less than 0.05) provides strong evidence against the null hypothesis (which states that the coefficient is zero, meaning no effect). In this case, several coefficients are significant with strong or very strong evidence against the null hypothesis, implying that those predictors have a significant relationship with the outcome.
How to Understand These Together
The R-squared and p-values are addressing different aspects of the model:

R-squared tells us about the overall fit of the model—how well the model explains the variance in the data.
P-values indicate the significance of individual predictor variables—whether they have a statistically significant effect on the outcome when other predictors are accounted for.
In this scenario:

The low R-squared value suggests that there are factors affecting the outcome that are not captured by the model. The model only accounts for a small portion of the total variability in the data.
Significant coefficients with strong evidence against the null hypothesis mean that some predictors still have statistically significant relationships with the outcome variable. This tells us that these predictors do have meaningful contributions, but they do not account for much of the overall variability.
Example Analogy
Imagine you're predicting house prices based on variables like the number of bedrooms and the size of the garden:

If the R-squared is low, it means that many factors determining house prices are not included in the model (e.g., neighborhood, age of the house, market conditions).
However, if the coefficient for the number of bedrooms is significant, it still means that the number of bedrooms affects the house price, but other important factors are missing, leading to a low overall explanatory power.

The model4_linear_form specification results in a complex design matrix (model4_spec.exog) with many new columns due to the inclusion of multiple interaction terms between variables (Attack, Defense, Speed, Legendary, Sp. How Model 4 Creates New Predictor Variables
The linear form (model4_linear_form) is specified to include both main effects (original variables) and interaction terms between the predictors.
The * operator is used in the formula, which means the model will include:
Each individual predictor, e.g., Attack, Defense, etc.
Two-way interactions, like Attack * Defense or Speed * Legendary.
Higher-order interactions, which involve three or more variables together, such as Attack * Defense * Speed.
This expansion of predictors creates many new columns in the design matrix (model4_spec.exog), each representing a different main effect or interaction term.
The design matrix essentially contains the numeric representations of these original variables and their interaction terms, used to make predictions for the outcome variable (model4_spec.endog).
Multicollinearity in the Design Matrix
Multicollinearity occurs when there is a high correlation between two or more predictors in the design matrix. This issue is prevalent in Model 4 due to the large number of interaction terms created, which tend to be highly correlated with each other or with the original variables.
You can observe this multicollinearity by computing the correlation matrix of the predictors (np.corrcoef(model4_spec.exog)). High values in this matrix indicate that many predictors are linearly dependent on others.
In the screenshot, we see that the condition number for Model 4 is extremely high (e.g., 12,000,000,000,000,000). The condition number is a measure of how sensitive the model is to changes in the input data. A very high condition number indicates severe multicollinearity, which causes:
Unstable Coefficients: The coefficients of the model can vary drastically with even small changes in the input data, leading to less reliable estimates.
Overfitting: With so many interrelated terms, the model can fit the training data almost perfectly, but it struggles to generalize to new, unseen data. This is because the model essentially "memorizes" the relationships in the training set, including the noise.
Lack of "Out of Sample" Generalization
The result of multicollinearity is that Model 4 has poor "out of sample" generalization. Specifically:
In training data, the model might perform very well, leading to a high in-sample R-squared value, because it has learned all the intricacies and even noise of the training set.
However, when applied to test data, it performs poorly, indicated by a low out-of-sample R-squared value. This is because the noise and redundant relationships captured by the complex interaction terms do not generalize well to new data.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import statsmodels.formula.api as smf

# Sample data frame to simulate (pokeaman data)
pokeaman = pd.DataFrame({
    'HP': np.random.randint(50, 150, 200),
    'Attack': np.random.randint(50, 150, 200),
    'Defense': np.random.randint(50, 150, 200),
    'Speed': np.random.randint(30, 120, 200),
    'Sp_Atk': np.random.randint(40, 130, 200),
    'Sp_Def': np.random.randint(40, 130, 200),
    'Legendary': np.random.choice([0, 1], 200),
})

# Lists to collect performance metrics
in_sample_r2_list = []
out_sample_r2_list = []

# Loop to create multiple train-test splits and fit models
for i in range(50):  # Perform 50 iterations
    # Split the data into train and test sets (different every time due to no fixed seed)
    pokeaman_train, pokeaman_test = train_test_split(pokeaman, train_size=0.5)
    
    # Define and fit the model on the training set
    model_spec = smf.ols(formula='HP ~ Attack + Defense + Speed + Sp_Atk + Sp_Def + Legendary', data=pokeaman_train)
    model_fit = model_spec.fit()
    
    # Calculate in-sample R-squared
    in_sample_r2 = model_fit.rsquared
    in_sample_r2_list.append(in_sample_r2)
    
    # Predict on the test set and calculate out-of-sample R-squared
    yhat_test = model_fit.predict(pokeaman_test)
    y_test = pokeaman_test['HP']
    out_sample_r2 = np.corrcoef(y_test, yhat_test)[0, 1] ** 2
    out_sample_r2_list.append(out_sample_r2)

# Visualization of the paired "in sample" and "out of sample" R-squared values
plt.figure(figsize=(12, 6))
plt.plot(in_sample_r2_list, label='In Sample R-squared', marker='o')
plt.plot(out_sample_r2_list, label='Out of Sample R-squared', marker='x')
plt.xlabel('Iteration')
plt.ylabel('R-squared')
plt.title('In Sample vs Out of Sample R-squared Over Multiple Iterations')
plt.legend()
plt.show()


Explanation of Results:
In-Sample R-squared: Measures model fit on training data, usually high since it minimizes training errors.
Out-of-Sample R-squared: Measures generalization on unseen data, often lower due to potential overfitting.

Purpose of Demonstration:
Variability of Performance: No fixed random seed shows how train-test splits impact model variability.
Model Stability: Consistently high in-sample vs. varying or low out-of-sample R-squared suggests overfitting.
Generalization Insight: Helps assess model reliability across different data splits for real-world application.
